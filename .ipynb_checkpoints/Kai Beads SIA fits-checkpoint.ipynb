```python
#TAMSD

def equation14(t, a, p, D, k):
    return 2.0*p - (D*sc.gamma(a+1))*(np.exp(-1.0*k/t))/(k**2) - (2*a*(a-1)*D)/((k**2) * (t**(2-a)))

ent_eq14=scipy.optimize.curve_fit(equation14,fullset[0],fullset[1], absolute_sigma=True)
print "ent is", ent_eq14[0]

micro_eq14=scipy.optimize.curve_fit(equation14,fullset[0],fullset[2], absolute_sigma=True)
print "micro is", micro_eq14[0]

act_eq14=scipy.optimize.curve_fit(equation14,fullset[0],fullset[3], absolute_sigma=True)
print "act is", act_eq14[0]

both_eq14=scipy.optimize.curve_fit(equation14,fullset[0][10:],fullset[4][10:], absolute_sigma=True)
print "both is", both_eq14[0]

coxl_eq14=scipy.optimize.curve_fit(equation14,fullset[0],fullset[5], absolute_sigma=True)
print "coxl is", coxl_eq14[0]
```

```python
#EAMSD

def equation11(t, plat, k, a, D):
    return plat - (2./k**2.)*a*(a-1.)*D*(t**(a-2.))*np.exp(-1.0*k*t)

ent_eq11=scipy.optimize.curve_fit(equation11,fullset[0],fullset[1], absolute_sigma=True)
print "ent is", ent_eq11[0]

micro_eq11=scipy.optimize.curve_fit(equation11,fullset[0][:],fullset[2][:], absolute_sigma=True)
print "micro is", micro_eq11[0]

act_eq11=scipy.optimize.curve_fit(equation11,fullset[0],fullset[3], absolute_sigma=True)
print "act is", act_eq11[0]

both_eq11=scipy.optimize.curve_fit(equation11,fullset[0][10:],fullset[4][10:], absolute_sigma=True)
print "both is", both_eq11[0]

coxl_eq11=scipy.optimize.curve_fit(equation11,fullset[0][50:],fullset[5][50:], absolute_sigma=True)
print "coxl is", coxl_eq11[0]
```

```python
#powerlaw

def powerlaw(t,k,a):
    return (k)*(t)**(a)

ent_pl=scipy.optimize.curve_fit(powerlaw,fullset[0][1:],fullset[1][1:], absolute_sigma=True)
print "ent is", ent_pl[0]

micro_pl=scipy.optimize.curve_fit(powerlaw,fullset[0][1:],fullset[2][1:], absolute_sigma=True)
print "micro is", micro_pl[0]

act_pl=scipy.optimize.curve_fit(powerlaw,fullset[0][1:],fullset[3][1:], absolute_sigma=True)
print "act is", act_pl[0]

both_pl=scipy.optimize.curve_fit(powerlaw,fullset[0][1:],fullset[4][1:], absolute_sigma=True)
print "both is", both_pl[0]

coxl_pl=scipy.optimize.curve_fit(powerlaw,fullset[0][1:],fullset[5][1:], absolute_sigma=True)
print "coxl is", coxl_pl[0]
```

```python
def equation2(t,k,g,b):
    return (1-np.exp(-k*t/g))/(b*k)

ent_eq2=scipy.optimize.curve_fit(equation2,fullset[0],fullset[1], absolute_sigma=True)
print "ent is", ent_eq2[0]

micro_eq2=scipy.optimize.curve_fit(equation2,fullset[0],fullset[2], absolute_sigma=True)
print "micro is", micro_eq2[0]

act_eq2=scipy.optimize.curve_fit(equation2,fullset[0],fullset[3], absolute_sigma=True)
print "act is", act_eq2[0]

both_eq2=scipy.optimize.curve_fit(equation2,fullset[0][:],fullset[4][:], absolute_sigma=True)
print "both is", both_eq2[0]

coxl_eq2=scipy.optimize.curve_fit(equation2,fullset[0][:],fullset[5][:], absolute_sigma=True)
print "coxl is", coxl_eq2[0]
```

```python
def equation13(t, a, D, k):
    return (2*D*(sc.gamma(a+1))/(k**a) + 2*D*(t**a)
           - (D/(k**a))*(np.exp(k*a)*sc.gammaincc(a+1.0, k*t) + np.exp(-1.0*k*t)*sc.gamma(a+1))
           -((k*D)/(a+1))*(t**(a+1))*np.exp(-1.0*k*t)*sc.hyp1f1(a + 1.0, a + 2.0, k*t, out=None)
           )

ent_eq13=scipy.optimize.curve_fit(equation13,fullset[0],fullset[1], absolute_sigma=True)
print "ent is", ent_eq13[0]

micro_eq13=scipy.optimize.curve_fit(equation13,fullset[0],fullset[2], absolute_sigma=True)
print "micro is", micro_eq13[0]

act_eq13=scipy.optimize.curve_fit(equation13,fullset[0],fullset[3], absolute_sigma=True)
print "act is", act_eq13[0]

both_eq13=scipy.optimize.curve_fit(equation13,fullset[0][:],fullset[4][:], absolute_sigma=True)
print "both is", both_eq13[0]

coxl_eq13=scipy.optimize.curve_fit(equation13,fullset[0][:],fullset[5][:], absolute_sigma=True)
print "coxl is", coxl_eq13[0]
```

```python
def equation9(t, a, D, k):
    return ((D/(k**a))*sc.gammainc(a+1.0, k*t) +2*D*(t**a)*np.exp(-1.0*k*1)
            -((k*D)/(a+1))*(t**(a+1))*np.exp(-2.0*k*t)*sc.hyp1f1(a + 1.0, a + 2.0, k*t, out=None)
           )

ent_eq9=scipy.optimize.curve_fit(equation9,fullset[0],fullset[1], absolute_sigma=True)
print "ent is", ent_eq9[0]

micro_eq9=scipy.optimize.curve_fit(equation9,fullset[0],fullset[2], absolute_sigma=True)
print "micro is", micro_eq9[0]

act_eq9=scipy.optimize.curve_fit(equation9,fullset[0],fullset[3], absolute_sigma=True)
print "act is", act_eq9[0]

both_eq9=scipy.optimize.curve_fit(equation9,fullset[0][:],fullset[4][:], absolute_sigma=True)
print "both is", both_eq9[0]

coxl_eq9=scipy.optimize.curve_fit(equation9,fullset[0][:],fullset[5][:], absolute_sigma=True)
print "coxl is", coxl_eq9[0]
```
